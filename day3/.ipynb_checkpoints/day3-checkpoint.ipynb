# 3. Speed optimization using Numpy
Revisit the `matmult.py` example from yesterday and improve its performance using Numpy.

```python
# Program to multiply two matrices using nested loops
import random
import time

start = time.time()

N = 250

# NxN matrix
X = []
for i in range(N):
    X.append([random.randint(0,100) for r in range(N)])

# Nx(N+1) matrix
Y = []
for i in range(N):
    Y.append([random.randint(0,100) for r in range(N+1)])

# result is Nx(N+1)
result = []
for i in range(N):
    result.append([0] * (N+1))

# iterate through rows of X
for i in range(len(X)):
    # iterate through columns of Y
    for j in range(len(Y[0])):
        # iterate through rows of Y
        for k in range(len(Y)):
            result[i][j] += X[i][k] * Y[k][j]

end = time.time()

print("Execution time: {:.2f} s.".format(end - start))

#for r in result:
#    print(r)
```

```python
# Program to multiply two matrices using nested loops
import random
import time
import numpy as np

start = time.time()

N = 250

# NxN matrix
X = np.random.randint(0, 100, (N, N))
#print("X is size {:}".format(X.shape))

# Nx(N+1) matrix
Y = np.random.randint(0, 100, (N, N+1))
#print("Y is size {:}".format(Y.shape))

result = np.matmul(X, Y)
print("result is size {:}".format(result.shape))

end = time.time()

print("Execution time: {:.2f} s.".format(end - start))

#for r in result:
#    print(r)
```
